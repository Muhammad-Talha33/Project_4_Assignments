### Objective:

### Accessing Elements:
### Write a function that:

### Accepts a list and an index as inputs.

### If the index is out of range, return an appropriate message.
### Modifying Elements:
### Write a function that:

### Accepts a list, an index, and a new value as inputs.
### Replaces the element at the specified index with the new value.
### If the index is out of range, return an appropriate message.
### Slicing the List:
### Write a function that:

### Accepts a list, a start index, and an end index as inputs.
### Returns a new list containing the elements from the start index up to (but not including) the end index.
### Handles cases where the indices are out of range.
### Game Interaction:
### Create a simple text-based game that:

### Prompts the user to select an operation (access, modify, slice).
### Asks for the necessary inputs (index, new value, etc.).
### Displays the result and the updated list.

```python
# For Accessing element
def access_element(lst, user_index):
 if user_index >= 0 and user_index < len(lst):
  return lst[user_index]
 else:
  return "Index out of range"

# For Modifying element

def modify_element(lst, user_index, new_value):
  if user_index >= 0 and user_index < len(lst):
    lst[user_index] = new_value
    return lst
  else:
    return "Index out of range"

# For Slicing Element
def slicing_ele(lst, start_index, end_index):
  if start_index >= 0 and end_index <= len(lst):
    return lst[start_index:end_index]
  else:
    return "Index out of range"
```

```python
def spaces():
  print(end="\n")

def index_game():
 lst = ["car", "bike", "bus", "train", "aeroplane"]
 user_choice = input("Which operation do you want to perform. (access, modify or slice) ")

 try:
  while True:
   if user_choice.lower() == "slice":
    spaces()
    print(f"Here's the current list: {lst}")
    start_index = int(input("Enter a starting index: "))
    end_index = int(input("Enter a ending index: "))
    sliced_result = slicing_ele(lst,start_index, end_index)
    print(f"Sliced list from index {start_index} to {end_index}: {sliced_result}")

   elif user_choice.lower() == "modify":
    spaces()
    print(f"Here's the current list: {lst}")
    user_index = int(input("Please enter the index you want to modify: "))
    new_value = input("enter a new value: ")
    updated_list = modify_element(lst, user_index, new_value)
    print(f"Element modified successfully! Updated list: {updated_list}")

   elif user_choice.lower() == "access":
    spaces()
    print(f"Here's the current list: {lst}")
    user_index = int(input("Enter an index: "))
    accessed_element = access_element(lst, user_index)
    print(f"Element at the specified index: {accessed_element}")

   else:
     print("Invalid operation. Please try again.")
     break
   spaces()

   user_choice = input("Which operation do you want to perform. (access, modify, slice or exit)? ")
   if user_choice.lower() == "exit":
    print("Exiting the program. Goodbye!")
    break

 except Exception as error:
   print(error)

index_game()
```
